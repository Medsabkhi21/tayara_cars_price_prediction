```python
df = pd.DataFrame()

page = soup(driver.page_source, 'html.parser')    # "page" contains the html code 
car_containers = page.find_all("div", class_="card")   # Find where the ad is located in the htm code. In this case it's in <div class="card">...</div>
print(len(car_containers),"ads has been found")    # len(car_containers) is the number of ads we loaded 
```

```python
for data in car_containers:
    find_link = data.find_all("a", class_="delta-link delta-link__primary")    # get the link of the ad. It's stored in <a class="delta-link delta-link__primary">...</a>
    L = str(find_link[0]).split(" ")          
    link = "https://tayara.tn"+L[3][6:-6]

    response2 = requests.get(link)      # Open the ad link
    page2 = soup(response2.text, 'html.parser')   # page2 contains the html code of the ad page 

    # Get the price
    try:
        car_containers2 = page2.find_all("div", class_="_3VhHF _1q9YR _3PJSU")[0]
        find_price = str(car_containers2.find_all("span")).split(" ")
        price_value = find_price[2].split(">")
        price_value = price_value[0][12:-1]
        price.append(price_value)
        print(price_value)
    except:  # In case if the ad doesen't containt a price we will add a NULL value
            #we will take care of price = 0.0dt later on
        price.append("NULL")
        print("NULL")
    
    # Get the location
    try:
        find_location = str(car_containers2.find_all("p", class_="_15IPb")[1])[46:-4]
        location.append(find_location)
        print(find_location)
    except:
        location.append("NULL")
        print("NULL")


    # Get the date
    try:
        find_date = str(car_containers2.find_all("p", class_="_15IPb")[0])[45:-11]
        date.append(find_date)
        print(find_date)
    except:
        date.append("NULL")
        print("NULL")

    # Get  km
    try:
        
        car_containers3 = page2.find_all("div",class_="DbXTC b7ygi")[3]
        #print(car_containers3)
        find_km = str(car_containers3.find_all("p", class_="_15IPb")[1])[18:-4]        
        print("\nkm : "+find_km)
        km.append(find_km)
    except:
        km.append("NULL")
        print('NULL')

    # Get annee
    try:
        find_annee = str(car_containers3.find_all("p", class_="_15IPb")[3])[18:-4]
        annee.append(find_annee)
        print(find_annee)
    except:
        annee.append("NULL")
        print("NULL")
    
    # Get marque
    try:
        find_marque = str(car_containers3.find_all("p", class_="_15IPb")[5])[18:-4]
        marque.append(find_marque)
        print(find_marque)
    except:
        marque.append("NULL")
        print("NULL")

    # Get model
    try:
        find_model = str(car_containers3.find_all("p", class_="_15IPb")[7])[18:-4]
        model.append(find_model)
        print(find_model)
    except:
        model.append("NULL")
        print("NULL")
        
        
     #get carburant   
    try:
        find_carburant = str(car_containers3.find_all("p", class_="_15IPb")[9])[18:-4]
        carburant.append(find_carburant)
        print(find_carburant)
    except:
        carburant.append("NULL")
        print("NULL")    
   
    #get puissance
    try:
        find_puissance = str(car_containers3.find_all("p", class_="_15IPb")[11])[18:-4]
        puissance.append(find_puissance)
        print(find_puissance)
    except:
        puissance.append("NULL")
        print("NULL")    
        
    #get boite
    try:
        find_boite = str(car_containers3.find_all("p", class_="_15IPb")[13])[18:-4]
        boite.append(find_boite)
        print(find_boite)
    except:
        boite.append("NULL")
        print("NULL")    
        
    #get cylindre
    try:
        find_cylindre = str(car_containers3.find_all("p", class_="_15IPb")[15])[18:-4]
        cylindre.append(find_cylindre)
        print(find_cylindre)
    except:
        cylindre.append("NULL")
        print("NULL")    
    
       
    print("---------------------")
   
    left=len(car_containers)-len(cylindre)
    print(str(left)+"times are left")
```

```python
df = pd.DataFrame()
df["location"] = location  
df["km"] = km
df["annee"] = annee
df["marque"] = marque
df["model"] = model
df["date"] = date
df["price"] = price
df["carburant"]=carburant
df["puissance"]=puissance
df["boite"]=boite
df["cylindre"]=cylindre
d = df.to_csv("./tayara_cars.csv")     # store the data in csv file

```
